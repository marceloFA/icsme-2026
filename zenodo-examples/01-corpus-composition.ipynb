# FixtureDB: Corpus Composition Analysis

This notebook provides an overview of the FixtureDB dataset, examining:
- **Language coverage**: How many repositories per language
- **Project maturity**: Star tier breakdown (popular vs. emerging)
- **Star distribution**: Spread of GitHub stars across languages
- **Pipeline status**: Repository collection and analysis success rates

These analyses establish the dataset scope and representativeness before deeper investigation.

## Setup: Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path

plt.rcParams['figure.facecolor'] = '#FAFAFA'
print("✓ Libraries imported successfully")
```

```python
data_dir = Path('fixturedb')
repos_path = data_dir / 'repositories.csv'
repos = pd.read_csv(repos_path)
print(f"✓ Loaded {len(repos)} repositories")
print(f"Shape: {repos.shape}")
repos.head()
```

## Language Distribution

```python
lang_counts = repos['language'].value_counts()
print(lang_counts)

fig, ax = plt.subplots(figsize=(10, 6), facecolor='#FAFAFA')
lang_counts.plot(kind='bar', ax=ax, color=sns.color_palette('husl', len(lang_counts)))
ax.set_title('Repositories by Language', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Language', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Star Tier Breakdown

```python
tier_by_lang = pd.crosstab(repos['language'], repos['star_tier'])
print(tier_by_lang)

fig, ax = plt.subplots(figsize=(10, 6), facecolor='#FAFAFA')
tier_by_lang.plot(kind='bar', stacked=True, ax=ax, color=['#2E86AB', '#A23B72'])
ax.set_title('Repositories by Language & Star Tier', fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Key Findings

**Dataset Scope:**
- FixtureDB contains 4 languages with syntax-based fixture detection
- Stratified by star tier (core vs. extended) for comparative analysis
- High collection success rate demonstrates robust pipeline
